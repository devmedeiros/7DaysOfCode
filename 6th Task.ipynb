# A/B Testing

The dataset comes from [Kaggle](https://www.kaggle.com/datasets/zhangluyuan/ab-testing/versions/1).

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../input/ab-testing/ab_data.csv')
data.head()
```

```python
print('Unique users: ', len(data.user_id.unique()), '\nRows in the data: ', len(data))
```

Let's remove users that are on both groups

```python
data.drop_duplicates(subset='user_id', keep='first', inplace=True)
```

Conversion rate

```python
data.groupby(['group']).converted.mean()*100.00
```

```python
g = sns.catplot(x="landing_page", y="converted", col="group",
                data=data, kind="bar", ci=None, estimator=sum)
```

```python
import scipy.stats as stats
```

```python
group1 = data[data.group == 'control'].converted
group2 = data[data.group == 'treatment'].converted
print('control size:', len(group1),'\ntreatment size:', len(group2))
```

**Ho:** The conversion is equal between the two groups

**Ha:** The conversion is not equal between the two groups

```python
stats.mannwhitneyu(group1, group2, alternative='two-sided')
```

This means we do not have sufficient evidence to say that the true conversion rate is different between the two groups.
